## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

### Challenge 1 – Probability with coloured balls

We have a box with:

* **10** white balls  
* **12** red balls  
* **8** black balls  

Total number of balls:



$$ 10 + 12 + 8 = 30 $$



---

#### 1. Probability of taking a white ball



$$ P(\text{white}) = \frac{10}{30} = \frac{1}{3} $$



---

#### 2. White first, black second, without replacement



$$ 
P(\text{white then black, no replacement}) 
= P(W_1)\cdot P(B_2 \mid W_1) 
= \frac{10}{30} \cdot \frac{8}{29} 
= \frac{80}{870} 
= \frac{8}{87} 
$$



---

#### 3. White first, black second, with replacement



$$
P(\text{white then black, with replacement})
= P(W_1)\cdot P(B_2)
= \frac{10}{30} \cdot \frac{8}{30}
= \frac{80}{900}
= \frac{4}{45}
$$



---

#### 4. Taking 3 balls, all different colours

We want one white, one red, and one black (order does not matter).

**Number of favourable ways** (1 of each colour):



$$
10 \cdot 12 \cdot 8 = 960
$$



**Total number of ways** to choose any 3 balls from 30:



$$
\binom{30}{3} = \frac{30 \cdot 29 \cdot 28}{6} = 4060
$$



So the probability is:



$$
P(\text{all different colours}) = \frac{960}{4060} = \frac{48}{203}
$$




## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

### Challenge 2 – Probability of rain on Saturday

We are given:

* 40% of all days are **cloudy**  
  

$$
  P(C) = 0.4
$$



* 50% of cloudy days are **rainy**  
  

$$
  P(R \mid C) = 0.5
$$



* 20% of non‑cloudy days are **rainy**  
  

$$
  P(R \mid C^c) = 0.2
$$



We want the **overall probability of rain**:



$$
P(R) = P(R \mid C)P(C) + P(R \mid C^c)P(C^c)
$$



Compute each part:



$$
P(C^c) = 1 - 0.4 = 0.6
$$





$$
P(R) = (0.5)(0.4) + (0.2)(0.6)
$$





$$
P(R) = 0.20 + 0.12 = 0.32
$$



#### **Final Answer**



$$
\boxed{P(\text{rain on Saturday}) = 0.32}
$$



There is a **32% chance of rain** next Saturday.


## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

### Challenge 3 – Checking probability statements

We are given:



$$
P(A) = 0.3,\quad P(B) = 0.6,\quad P(A \cap B) = 0.1
$$



We will check each statement.

---

#### 1. $ P(\overline{A}) = 0.7 $



$$
P(\overline{A}) = 1 - P(A) = 1 - 0.3 = 0.7
$$



**True**

---

#### 2. $ P(A \cup B) = 0.6 $



$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
= 0.3 + 0.6 - 0.1 = 0.8
$$



So $ P(A \cup B) = 0.8 \neq 0.6 $.

**False**

---

#### 3. $ P(A \cap \overline{B}) = 0.2 $



$$
P(A \cap \overline{B}) = P(A) - P(A \cap B) = 0.3 - 0.1 = 0.2
$$



**True**

---

#### 4. $ P(A \cup \overline{A}) = 1 $

Since $ A \cup \overline{A} $ is the whole sample space:



$$
P(A \cup \overline{A}) = 1
$$



**True**

---

#### 5. $ P(A \cap (B \cup \overline{B})) = 0.4 $

But $ B \cup \overline{B} $ is the whole sample space, so:



$$
A \cap (B \cup \overline{B}) = A
\Rightarrow P(A \cap (B \cup \overline{B})) = P(A) = 0.3
$$



So it is not 0.4.

**False**

---

#### 6. $ P(\overline{A} \cap \overline{B}) = 0.3 $



$$
P(\overline{A} \cap \overline{B}) = 1 - P(A \cup B) = 1 - 0.8 = 0.2
$$



So it is not 0.3.

**False**

---

#### 7. $ P(\overline{A \cap B}) = 0.9 $



$$
P(\overline{A \cap B}) = 1 - P(A \cap B) = 1 - 0.1 = 0.9
$$



**True**

---

#### Summary

* $ P(\overline{A}) = 0.7 $ →  True  
* $ P(A \cup B) = 0.6 $ →  False  
* $ P(A \cap \overline{B}) = 0.2 $ →  True  
* $ P(A \cup \overline{A}) = 1 $ →  True  
* $ P(A \cap (B \cup \overline{B})) = 0.4 $ →  False  
* $ P(\overline{A} \cap \overline{B}) = 0.3 $ →  False  
* $ P(\overline{A \cap B}) = 0.9 $ →  True


## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

### Challenge 4 – Probability of having flown to China given that you got Covid

We are given:

* If you fly to **China**, probability of Covid:  
  

$$
  P(Covid \mid China) = 0.01
$$



* If you fly to **Italy**, probability of Covid:  
  

$$
  P(Covid \mid Italy) = 0.20
$$



* Dice rule:  
  * If the roll is **1**, you fly to Italy  
    

$$
    P(Italy) = \frac{1}{6}
$$


  * Otherwise, you fly to China  
    

$$
    P(China) = \frac{5}{6}
$$



We want:



$$
P(China \mid Covid)
$$



This is a **Bayes' theorem** problem:



$$
P(China \mid Covid) = 
\frac{P(Covid \mid China)P(China)}
     {P(Covid \mid China)P(China) + P(Covid \mid Italy)P(Italy)}
$$



---

#### Step-by-step computation



$$
P(Covid \mid China)P(China) = 0.01 \cdot \frac{5}{6} = \frac{0.05}{6}
$$





$$
P(Covid \mid Italy)P(Italy) = 0.20 \cdot \frac{1}{6} = \frac{0.20}{6}
$$



Total probability of Covid:



$$
P(Covid) = \frac{0.05}{6} + \frac{0.20}{6} = \frac{0.25}{6}
$$



Now apply Bayes:



$$
P(China \mid Covid) = 
\frac{\frac{0.05}{6}}{\frac{0.25}{6}}
= \frac{0.05}{0.25}
= 0.2
$$



---

### **Final Answer**



$$
\boxed{P(\text{flew to China} \mid \text{got Covid}) = 0.2}
$$



There is a **20% chance** you flew to China given that you got Covid.

### BONUS – Sum of two dice: 7, 8, 9 and simulation

We roll **2 fair dice**. Each die has outcomes \(1,2,3,4,5,6\).  
Total number of possible outcomes:



$$
6 \times 6 = 36
$$



---

#### 1. Probability that the sum is 7

Favourable pairs \((d_1, d_2)\):



$$
(1,6), (2,5), (3,4), (4,3), (5,2), (6,1)
$$



There are \(6\) such outcomes.



$$
P(\text{sum} = 7) = \frac{6}{36} = \frac{1}{6}
$$



---

#### 2. Probability that the sum is 8

Favourable pairs:



$$
(2,6), (3,5), (4,4), (5,3), (6,2)
$$



There are \(5\) such outcomes.



$$
P(\text{sum} = 8) = \frac{5}{36}
$$



---

#### 3. Probability that the sum is 9

Favourable pairs:



$$
(3,6), (4,5), (5,4), (6,3)
$$



There are \(4\) such outcomes.



$$
P(\text{sum} = 9) = \frac{4}{36} = \frac{1}{9}
$$



---

#### 4. Simulation of 10 000 rolls and countplot


```python
import random
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# simulate 10 000 rolls of two dice
n = 10000
sums = []

for _ in range(n):
    d1 = random.randint(1, 6)
    d2 = random.randint(1, 6)
    sums.append(d1 + d2)

# put results in a DataFrame
df = pd.DataFrame({"sum": sums})

# draw a countplot of the sums
plt.figure(figsize=(8, 4))
sns.countplot(x="sum", data=df)
plt.title("Distribution of sums of two dice (10 000 simulations)")
plt.show()
```
